```python
import sqlite3
import csv
import os
from openai import OpenAI

client = OpenAI(api_key = os.getenv("GOOGLE_API_KEY"))

#DBファイル作成
if os.path.exists("club.db"):
    os.remove("club.db")
conn = sqlite3.connect('./club.db')
cur = conn.cursor()

#データの整合性保証(外部キー制約ON)
cur.execute('PRAGMA foreign_keys = ON')



#ポジションリストにcsvファイルからデータを入れる
cur.execute("""
    CREATE TABLE position(
        id_posi  INTEGER PRIMARY KEY,
        position TEXT
    )
""")

with open('position.csv', 'r', encoding = 'utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        cur.execute(
            'INSERT INTO position(id_posi, position) VALUES(?, ?)',
            (row[0], row[1])
        )

#データ確認用
# cur.execute('SELECT * FROM position')
# rows = cur.fetchall()
# for row in rows:
#     print(f"ポジション番号:{row[0]}, ポジション:{row[1]}")


#選手リストにcsvファイルからデータを入れる
cur.execute("""
    CREATE TABLE player(
        id_pl   TEXT PRIMARY KEY,
        name    TEXT,
        id_posi INTEGER REFERENCES position(id_posi)
    )
""")

with open('player.csv', 'r', encoding = 'utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        cur.execute(
            'INSERT INTO player(id_pl, name, id_posi) VALUES(?, ?, ?)',
            (row[0], row[1], row[2])
        )

#データ確認用
# cur.execute('SELECT * FROM player')
# rows = cur.fetchall()
# for row in rows:
#     print(f"選手ID:{row[0]}, 名前:{row[1]}, ポジション番号{row[2]}")


#メニューリストにcsvファイルからデータを入れる
cur.execute("""
    CREATE TABLE menu(
        id_m      TEXT PRIMARY KEY,
        menu_name TEXT,
        id_posi   INTEGER REFERENCES position(id_posi),
        memo      TEXT
    )
""")

with open('menu.csv', 'r', encoding = 'utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        cur.execute(
            'INSERT INTO menu(id_m, menu_name, id_posi, memo) VALUES(?, ?, ?, ?)',
            (row[0], row[1], row[2], row[3])
        )

#データ確認用
# cur.execute('SELECT * FROM menu')
# rows = cur.fetchall()
# for row in rows:
#     print(f"メニューID:{row[0]}, メニュー名:{row[1]}, ポジション番号:{row[2]}, コメント:{row[3]}")


conn.commit()
conn.close()
```
